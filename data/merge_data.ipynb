```python
ver_0 = pd.read_csv('wisdom2eg_0.tsv', sep='\t')
ver_0
```

```python
ver_1 = pd.read_csv('egs_koreaUniv_v0.tsv', sep='\t')
ver_1 = ver_1[['wisdom', 'eg']]
ver_1
```

```python
# there may be better way
for i in range(10):
    wisdom = ver_0.iloc[i]['wisdom']
    ver_0_egs = ver_0.iloc[i].filter(regex='eg_*')
    ver_0_not_nan_egs = ver_0_egs[ver_0_egs.notnull()]
    for eg in ver_0_not_nan_egs:
        ver_1 = ver_1.append({'wisdom': wisdom, 'eg': eg}, ignore_index=True)
```

```python
ver_1.sort_values(axis=0, inplace=True, by='wisdom')
ver_1.reset_index(drop=True, inplace=True)
```
